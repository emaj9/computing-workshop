```python
import numpy as np

%matplotlib notebook

w_1 = np.array([1, 2, 2])
b_1 = 4
w_2 = np.array([3, 2, 3])
b_2 = -3

# The sigmoid function. We use this as an activation function.
# Used for feedforward.
def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))

def evaluate(a):
    return (sigmoid(np.dot(w_1, a) + b_1), sigmoid(np.dot(w_2, a) + b_2))

X = []
for x in range(5):
    for y in range(5):
        for z in range(5):
            X.append( np.array([x, y, z]) )

Y = [evaluate(x) for x in X]

from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(9, 8))
ax = fig.add_subplot(111, projection='3d')

OUT_X, OUT_Y, OUT_Z = zip(*[x for x, y in zip(X, Y) if y[0] > y[1]])
IN_X, IN_Y, IN_Z = zip(*[x for x, y in zip(X, Y) if y[0] <= y[1]])

ax.scatter(OUT_X, OUT_Y, OUT_Z, c='r', marker='o')
ax.scatter(IN_X, IN_Y, IN_Z, c='b', marker='^')
```
