```python
import pandas as pd
```

```python
df = pd.read_csv('colors.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df.red, df.green, df.blue)
    
```

```python
from sklearn.cluster import KMeans
```

```python
scanner = KMeans(n_clusters=8)
```

```python
scanner.fit(df[['red', 'green', 'blue']])
```

```python
scanner.labels_
```

```python
df['cluster'] = scanner.labels_
```

```python
print(df.to_string())
```

```python
from sklearn.cluster import DBSCAN
```

```python
scanner = DBSCAN(eps=25, min_samples=3)
```

```python
scanner.fit(df[['red', 'green', 'blue']])
```

```python
scanner.labels_
```

```python
df['dbscan_clusters'] = scanner.labels_
```

```python
print(df.to_string())
```
