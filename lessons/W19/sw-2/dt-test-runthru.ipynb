```python
import matplotlib.pyplot as plt
%matplotlib inline
both = list(zip(((x[0], x[2]) for x in dataset['data']), dataset['target']))
for i, mark in zip(set(dataset['target']), '^ov+'):
    X, Y = list(zip(*[p for p, d in both if d == i]))
    plt.scatter(X, Y, marker=mark)
plt.show()
```
